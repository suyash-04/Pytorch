```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['Unnamed: 32', 'id'], inplace=True)
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2, random_state=42, stratify=df.iloc[:, 0])
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

```python
#Defining the model
class MySimpleNN():
    def __init__(self, X):
        self.weights = torch.randn(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
        self.bias = torch.zeros(1,dtype= torch.float64, requires_grad=True)
    
    def forward(self, X):
        z= torch.matmul(X, self.weights) + self.bias
        return torch.sigmoid(z)
    
    def loss(self, y_pred, y_true):
        return torch.mean(-y_true * torch.log(y_pred) - (1 - y_true) * torch.log(1 - y_pred))
    
```

```python

#imp parameters
learning_rate = 0.76
epochs = 1000
```

```python
#Training Pipeline 
model = MySimpleNN(X_train_tensor)

for epoch in range(epochs):
#forward pass
    y_pred = model.forward(X_train_tensor)
    
#loss calculate 
    loss = model.loss(y_pred, y_train_tensor)
    
#backward pass 
    loss.backward()

#updating params
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad

#zeroing gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()
#Printing loss
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```

```python
#Evalution
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.7).int()
     
```

```python
accuracy = (y_pred == y_test_tensor).float().mean()
```
